```python
import requests
import boto3
import psycopg2
import os
import time
```

```python
s3 = boto3.client('s3')
```

```python
url = 'https://api.scryfall.com/cards/named?exact=delver of secrets&format=img&version=png'
```

```python
response = requests.get(url)

response.status_code
```

```python
type(response.content)
```

```python
s3.put_object(Bucket='mtg-capstone', Key='card_images/2259.png', Body=response.content)
```

```python
hostname = os.environ['CAPSTONE_DB_HOST']
dbname = os.environ['CAPSTONE_DB_DBNAME']
username = os.environ['CAPSTONE_DB_USERNAME']
password = os.environ['CAPSTONE_DB_PASSWORD']
```

```python
conn = psycopg2.connect('dbname={} host={} user={}'.format(dbname, hostname, username))
```

```python
cursor = conn.cursor()
```

```python
cursor.execute(query='SELECT scryfall_id, cardstorm_id FROM cards')
```

```python
results = cursor.fetchall()
```

```python
results
```

```python
for scryfall_id, cardstorm_id in results:
    url = 'https://api.scryfall.com/cards/{}?format=image&'.format(scryfall_id)
    response = requests.get(url)
    
    s3.put_object(Bucket='mtg-capstone', Key='card_images/jpg/{}.jpg'.format(cardstorm_id), Body=response.content)
    
    time.sleep(0.1)
```
